# 4. 利用OpenCV实现实时物体识别

```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
# 调用摄像头
cap=cv2.VideoCapture(0) #默认摄像头开启

while cap.isOpened():
    ret, frame = cap.read()
    results = model(frame)
    cv2.imshow('YOLO', np.squeeze(results.render()))
    if cv2.waitKey(10) & 0xFF==27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# 使用视频测试
cap=cv2.VideoCapture('./video/3.mp4') #默认摄像头开启

while cap.isOpened():
    ret, frame = cap.read()
    results = model(frame)
    cv2.imshow('YOLO', np.squeeze(results.render()))
    if cv2.waitKey(10) & 0xFF==27:
        break

cap.release()
cv2.destroyAllWindows()
```

# 5.训练定制模型

导包

```python
import uuid
import os
import time
```

参数配置

```python
IMAGE_PATH=os.path.join('data', 'images')
labels=['me', 'me-no-glasses']
num_imgs=10
```

采集图像数据集

```python
cap=cv2.VideoCapture(0) #默认摄像头0开启

for label in labels:
    print(f'--------- \n collecting images for {lable} \n ---------')
    time.sleep(10) #每类采集之前缓冲10s
    
    for i in range(num_imgs): #每一类采集20张
        print(f'collecting {label}.{i+1}')
        
        ret, frame = cap.read()
        
        img_name=os.path.join(IMAGE_PATH, label + '.' + str(uuid.uuid1()) + '.jpg')        
        cv2.imwrite(img_name, frame)
        cv2.imshow('Data Collecting...', frame)
        time.sleep(2)
        if cv2.waitKey(10) & 0xFF==27:
            break

cap.release()
cv2.destroyAllWindows()
```
